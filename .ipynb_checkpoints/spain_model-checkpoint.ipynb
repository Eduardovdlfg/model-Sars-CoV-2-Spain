```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import csv
from scipy import signal
import numpy as np
```

```python
spain = pd.read_csv('covid.csv')
```

```python
spain=spain.loc[spain["location"]=="Spain"]
spain.index = range(len(spain))
spain
```

dia 186 é o sábado que começou a ter problemas ou 2020-07-03

```python
#conserta=spain.loc[spain['date']>='2020-07-03']
new_cases=list(spain['new_cases'])
new_deaths=list(spain["new_deaths"])
new_cases=new_cases[0:-3]
new_deaths=new_deaths[0:-3]
total_cases=[0]
total_deaths=[0]

new_cases = signal.savgol_filter(new_cases,7,1)
new_deaths= signal.savgol_filter(new_deaths,7,1)
for i in range(1,len(new_cases)):
    total_cases.append(total_cases[i-1]+new_cases[i])   
    total_deaths.append(total_deaths[i-1]+new_deaths[i])
```

```python
fig, axs = plt.subplots(2,1, sharex=True,figsize=(10, 10))
fig.suptitle('Vertically stacked subplots')
axs[0].plot(range(len(new_cases)),new_cases, label='Casos diários')
axs[0].plot(range(len(new_cases)),new_deaths,label='Mortes diárias')
axs[0].grid(True)
axs[0].legend()
axs[1].plot(range(len(new_cases)),total_cases, label='Casos totais')
axs[1].plot(range(len(new_cases)),total_deaths,label='Mortes totais' )
axs[1].grid(True)
axs[1].legend()
```
