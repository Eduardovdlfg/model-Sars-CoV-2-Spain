```python
from lmfit import minimize, Parameters, Parameter, report_fit
def seiru (x0,t, ps):
    
    try:
        tau = ps['tau'].value
        alpha = ps['alpha'].value
        eta = ps['eta'].value
        v1 = ps['v1'].value
        v = ps['v'].value
        
    except:
        tau, alpha, eta, v1, v = ps
        
    
    S,E,I,R,U = x0
    
    return [-tau*S*(I + U),         #dS/dt
            tau*S*(I + U) -alpha*E, #dE/dt
            alpha*E-v*I,            #dI/dt
            v1*I-eta*R,             #dR/dt
            (v-v1)*I-eta*U          #dU/dt

def engine(t, x0, ps):
    St = 1 - np.sum(x0)
    x0 = np.insert(x0,0,St)
    x = odeint(seiru,x0, t, args = (ps,))
    return x

def res(ps, ts, data):
    x0 = ps['E'].value, ps['I'].value, ps['R'].value, ps['U'].value
    model = engine(ts, x0, ps)
    
    #teste
    return (model[:,2] - data).ravel()      
```

```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import odeint
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "notebook"
%matplotlib inline
plt.style.use('ggplot')
from matplotlib.widgets import Slider, Button, RadioButtons
from IPython.html.widgets import *
```

```python
t = np.linspace(0,251,251)
x0 = np.array([0.2,0,0,0])

tau = 0.3
alpha = 0.2
eta = 0.1
v1 = 0.5
v = 0.5


t_params = np.array((tau, alpha, eta, v1, v))
```

```python
sol = engine(t,x0,t_params)

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

for i in range(5):
    plt.plot(sol[:,i])
    
plt.show()
```

```python
data = sol[:,2] 
```

```python
params = Parameters()

params.add('E', value=0.0, min=0, max=1,vary= False )
params.add('I', value=0.1, min=0, max=1, vary= False)
params.add('R', value=0.0, min=0, max=1, vary= False)
params.add('U', value=0.0, min=0, max=1, vary= False)

params.add('tau', value=0.2, min=0, max=1)
params.add('alpha', value=0.5, min=0, max=1)
params.add('eta', value=0.5, min=0, max=1)
params.add('v1', value=0.5, min=0, max=1)
params.add('v', value=0.5, min=0, max=1)



result = minimize(res, params, args = (t, data), method= 'leastsq')
```

```python
final = data + result.residual.reshape(data.shape)
plt.figure(num=None, figsize=(15, 7), dpi=80, facecolor='w', edgecolor='k')
plt.plot(t, data, 'o')
plt.plot(t, final, '-', linewidth=2)
```
